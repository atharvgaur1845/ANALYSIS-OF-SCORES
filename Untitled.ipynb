```python
import pandas as pd

# Load the data from the CSV file
df = pd.read_csv('meow-total-randomised-noise.csv')

# Replace 'woof' with 0
df.replace('woof', 0, inplace=True)

# Convert columns to numeric types
numeric_columns = ['T1', 'T2', 'T3', 'T4', 'T5', 'Midsem', 'Compre A', 'Compre B']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric)

# Function to calculate the sum of the best 4 out of 5 scores
def best_4_of_5(row):
    scores = [row['T1'], row['T2'], row['T3'], row['T4'], row['T5']]
    scores.sort(reverse=True)
    return sum(scores[:4])

df['Best_4_Tutorials'] = df.apply(best_4_of_5, axis=1)

# Calculate total marks
df['Total_Marks'] = df['Best_4_Tutorials'] + df['Midsem'] + df['Compre A'] + df['Compre B']

# Get basic statistics
stats = df.describe()
data_types = df.dtypes
print(stats)
print(data_types)

import matplotlib.pyplot as plt

# Plot distribution of total marks
plt.figure(figsize=(12, 6))
plt.hist(df['Total_Marks'], bins=30, edgecolor='black', alpha=0.7)

# Calculate and plot mean, median, and standard deviation
mean = df['Total_Marks'].mean()
median = df['Total_Marks'].median()
std_dev = df['Total_Marks'].std()
q1 = df['Total_Marks'].quantile(0.25)
q3 = df['Total_Marks'].quantile(0.75)

plt.axvline(mean, color='r', linestyle='dashed', linewidth=1, label=f'Mean: {mean:.2f}')
plt.axvline(median, color='g', linestyle='dashed', linewidth=1, label=f'Median: {median:.2f}')
plt.axvline(mean - std_dev, color='b', linestyle='dotted', linewidth=1, label=f'-1 Std Dev: {mean - std_dev:.2f}')
plt.axvline(mean + std_dev, color='b', linestyle='dotted', linewidth=1, label=f'+1 Std Dev: {mean + std_dev:.2f}')
plt.axvline(q1, color='purple', linestyle='dashdot', linewidth=1, label=f'1st Quartile: {q1:.2f}')
plt.axvline(q3, color='purple', linestyle='dashdot', linewidth=1, label=f'3rd Quartile: {q3:.2f}')

plt.title('Distribution of Total Marks')
plt.xlabel('Total Marks')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()


# Extract the branch from the ID column
df['Branch'] = df['ID'].str[4:6]

# Group by branch and calculate frequency distribution of Total_Marks
grouped = df.groupby('Branch')['Total_Marks']

# Create histograms for each branch
fig, axes = plt.subplots(nrows=len(grouped.groups), figsize=(8, 6*len(grouped.groups)))
for i, (branch, group) in enumerate(grouped.groups.items()):
    axes[i].hist(group, bins=10, alpha=0.7, color='skyblue', edgecolor='black')
    axes[i].set_title(f"Branch {branch} - Total Marks Distribution")
    axes[i].set_xlabel("Total Marks")
    axes[i].set_ylabel("Frequency")

# Layout so plots do not overlap
fig.tight_layout()




plt.title('Distribution of Total Marks by Branch')
plt.xlabel('Total Marks')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

```
